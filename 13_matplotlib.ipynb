## Adding Style to our graph

```python

from matplotlib import style

style.use("ggplot")


x=[5,8,10]
y= [12,16,4]

x2= [6,9,11]
y2= [6,15,7]

plt.plot(x,y,'g',label='LineOne', linewidth=5)
plt.plot(x2,y2,'r',label='LineTwo', linewidth=5)

plt.title("Style Graph")
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.legend()
plt.grid(True, color='k')
plt.show()
```

```python

plt.bar([1,3,5,7,9],[5,2,7,8,2], label="Example One",color='r')
plt.bar([2,4,6,8,10],[8,6,2,5,6], label="Example Two", color='g')
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar Graph')

plt.show()
```
